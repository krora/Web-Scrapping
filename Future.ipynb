```python
import bs4 as bs
import urllib.request

source = urllib.request.urlopen('http://datascience.community/colleges').read()
```

```python
soup = bs.BeautifulSoup(source,'lxml')
```

```python
table = soup.table
```

```python
table = soup.find('table')
```

```python
table_rows = table.find_all('tr')
```

```python
print(table_rows)
```

```python
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```

```python
import pandas as pd

dfs = pd.read_html('http://datascience.community/colleges',header=0)
for df in dfs:
    print(df)
```
